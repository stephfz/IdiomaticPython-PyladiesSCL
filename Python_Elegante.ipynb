### 4. Valores Truthy y Falsy
Debemos evitar la comprobacion verdadero/falso son valores booleanos. Es decir, usar siempre valores booleanos intrínsecos

````python
# Nos vemos tentadxs a hacer esto
x = True
y = 0

if x == True:
  print("Verdadero")
elif x == False:
  print("Falso")

if y == 0:
  print("valor cero")

ls = [2,5]
if len(ls) > 0:
  print("tiene valores")  
````

````python
# Pero hay una manera más elegante:
x = True
y = 0

if x:
  print("Verdadero")
else:
  print("Falso")

if not y:
  print("valor cero")

# valores intrinsecos de OBJETOS python
ls = [2,5]
#ls = []
if ls:
  print("tiene valores") 
````

### 5. Operador ternario

Python no tiene operador ternario por defecto pero hay una alternativa.

Ej. Java


```
res=(num1>num2) ? (num1+num2):(num1-num2)
```



````python
# Sin conocer la alternativa al operador ternario 

num1 = 20
num2 = 5

res  = 0
is_greater = num1 > num2

if is_greater:
  res = num1 - num2
else:
  res = num1 + num2  

print(res)  
````

````python
# Conociendo la alternativa ;)
is_greater = num1 > num2
res = (num1 - num2) if is_greater else (num1 + num2)

print(res)
````

### 10. Comprensión de listas
En la medida de lo posible, crear listas en una sola linea para mejor legibilidad.
[Ref](https://ellibrodepython.com/list-comprehension-python)

````python
prime_numbers = list()

for element in range(10):
  if (element % 2):
    prime_numbers.append(element)

print(prime_numbers)    

# Podemos hacer uso de funciones lambda
prime_numbers = list(filter(lambda element: not(element % 2), range(10)))
#print(filter(lambda element: not(element % 2), range(10)))
print (prime_numbers)
````

````python
# Un poco mejor, usando funciones lambda
prime_numbers_lambda = list(filter(lambda element: element % 2, range(10)))
#print(filter(lambda element: element % 2, range(10)))
print (prime_numbers_lambda)
````

````python
# Usando compresion
prime_numbers_comp = [element for element in range(10) if (element % 2)]

print (prime_numbers_comp)
````

### 11. `enumerate` para Indices y Elementos

````python
# Usando variables innecesarias

the_weekend = ['viernes', 'sabado', 'domingo']
index = 0
while index < len(the_weekend):
  print(the_weekend[index], index)
  index += 1
````

````python
#the_weekend = ['viernes', 'sabado', 'domingo']
for index, value in enumerate(the_weekend):
  print(value, index)

# otra forma
print (list(enumerate(the_weekend)))

````

### 12. Comprensión de Diccionarios
A veces, por alguna motivo queremos "cruzar" listas para crear un diccionario.

````python
# Mal

headers = ['ciudad', 'densidad', 'poblacion']
data = ['Narnia', 45.6, 450]

dicc = {}
for header in headers:
  for dato in data:
    dicc[header] = dato


print(dicc)    
````

````python
# Gud
my_shiny_dict = {i:j for i,j in zip(headers, data)}

print(my_shiny_dict)  
````

### 13. Compresión de sets
Similar a la compresion de listas, solo que reemplazamos le `( )` por `{ }`

````python
# Sin compresion
frase = "La mar estaba serena, serena estaba la mar"
unique_elements = set()

unique_elements = set()
for word in frase:
  for char in word:
    unique_elements.add(char)

print(unique_elements)
````

````python
# Compresion  == one liner :)
unique_elements = {i for i in frase}
print (unique_elements)
````

### 15. Swapping

````python
# Si no existiera swaping
a = 5
b = 10
temp = a
a = b
b = temp

print(f"a: {a}, b: {b}")

# con swaping
a = 5
b = 10

a,b = b,a
print(f"a: {a}, b: {b}")
````
